```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors
```

```python
url = 'https://github.com/ArinB/CA05-kNN/raw/master/movies_recommendation_data.csv'
df = pd.read_csv(url)
df.head()
```

```python
# drop label column
df = df.drop(['Label'], axis =1)
df.info()
```

```python
df_movies = df.iloc[:,0:2]
df_movies.head()
```

```python
df_features = df.drop(['Movie Name'], axis =1)
df_features.head()
```

```python
# create a dictionary for movies and their indices
movie_dict = {}
for i in range(0,30):
  movie_dict[i] = df_movies.iloc[i,1]
movie_dict
```

# Building KNN Model

https://medium.com/towards-data-science/knn-using-scikit-learn-c6bed765be75

```python
X = df_features.loc[:, df_features.columns != 'Movie ID']
# y = df_features.loc[:, df_features.columns == 'Movie ID']
X.head()
```

```python
# innitialize model
knn = NearestNeighbors(n_neighbors=5)
# fit model
knn.fit(X)
```

```python
# get indices of the recommendations
distance, indices = knn.kneighbors([[7.2, 1, 1, 0, 0, 0, 0, 1]])
print(distance)
print(indices)
```

```python
# get movie names 
def get_rec():
  print('Here are the top 5 recommended movies for you:')
  for i in indices[0]:
    print(movie_dict[i])
```
